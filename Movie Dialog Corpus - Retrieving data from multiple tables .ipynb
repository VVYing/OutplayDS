```python
import pandas as pd
lines = pd.read_csv('C:\Kaggle\DataSet\Movie Dialog Corpus\movie_lines.tsv', sep='\t', error_bad_lines=False, warn_bad_lines=False, header=None)
lines.columns = ['LineID', 'CharacterID', 'MovieID', 'CharacterName', 'LineText']
lines.head()
```

```python
lines.shape
```

```python
characters = pd.read_csv('C:\Kaggle\DataSet\Movie Dialog Corpus\movie_characters_metadata.tsv', sep='\t', error_bad_lines=False, warn_bad_lines=False, header=None)
characters.columns = ['CharacterID', 'CharacterName', 'MovieID', 'MovieTitle', 'Gender', 'PositionCredits']
characters.head()
```

```python
characters.shape
```

```python
df = pd.merge(lines, characters, how='inner', on=['CharacterID','MovieID', 'CharacterName'],
         left_index=False, right_index=False, sort=True,
         copy=False, indicator=False)
df.head()
```

```python
movies = pd.read_csv("C:\Kaggle\DataSet\Movie Dialog Corpus\movie_titles_metadata.tsv", sep='\t', error_bad_lines=False, warn_bad_lines=False, header=None)
movies.columns = ['MovieID','MovieTitle','ReleaseYear','Rating','Votes','Genres']
movies.head()
```

```python
df = pd.merge(df, movies, how='inner', on=['MovieID','MovieTitle'], left_index=False, right_index=False, sort=True, copy=False, indicator=False)
df.head()
```
